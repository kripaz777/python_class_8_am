```python
info = ""
file = open("info.csv",'w')
file.write("name,age,add\n")
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    age = int(input("Enter age = "))
    add = input("Enter add = ")
    info = info + f"{name},{age},{add}\n"
    
file.write(info)
file.close()
```
